```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
points = [(3.5, 4), (1, 3), (0, 0)]

polar_points = [(np.sqrt(x**2 + y**2), np.arctan2(y, x)) for x, y in points]

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

for r, theta in polar_points:
    ax.scatter(theta, r)

ax.set_ylim(0, 5) 
ax.set_yticklabels([])

ax.spines['polar'].set_visible(False)

for point, label in zip(polar_points, ['(3.5, 4)', '(1, 3)', '(0, 0)']):
    plt.text(point[1], point[0], label, fontsize=12, ha='left')

plt.savefig("imgs/3_9.png")
```
