```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, ax = plt.subplots()

plt.xlim(-2, 3)
plt.ylim(-2, 2)

points = [(-1, -1), (0, 0), (2, 1)]
for point in points:
    ax.plot(*point, marker="o", color="blue")

ax.text(-1, -1, "(-1, -1)", verticalalignment="top", horizontalalignment="right")
ax.text(0, 0.2, "(0, 0)", verticalalignment="top", horizontalalignment="right")
ax.text(2, 1, "(2, 1)", verticalalignment="bottom", horizontalalignment="left")

ax.axhline(0, color="black", linewidth=0.5) 
ax.axvline(0, color="black", linewidth=0.5)

ax.plot([2, 2], [0, 1], color="black", linestyle="dashed", lw = 1)
ax.plot([0, 2], [1, 1], color="black", linestyle="dashed", lw = 1)
ax.plot([-1, -1], [0, -1], color="black", linestyle="dashed", lw = 1)
ax.plot([0, -1], [-1, -1], color="black", linestyle="dashed", lw = 1)

ax.text(2.1, 0.5, "y = 1", verticalalignment="center", horizontalalignment="left", rotation=0)
ax.text(1, 1, "x = 2", verticalalignment="bottom", horizontalalignment="center", rotation=0)
ax.text(-1.1, -0.5, "y = -1", verticalalignment="center", horizontalalignment="right", rotation=0)
ax.text(-0.25, -1, "x = -1", verticalalignment="top", horizontalalignment="right", rotation=0)

ax.grid(True)
plt.savefig("imgs/3_1.png")
plt.show()
```

```python
points = [(3.5, 4), (1, 3), (0, 0)]

polar_points = [(np.sqrt(x**2 + y**2), np.arctan2(y, x)) for x, y in points]

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

for r, theta in polar_points:
    ax.scatter(theta, r)

ax.set_ylim(0, 5) 
ax.set_yticklabels([])

ax.spines["polar"].set_visible(False)

for point, label in zip(polar_points, ["(3.5, 4)", "(1, 3)", "(0, 0)"]):
    plt.text(point[1], point[0], label, fontsize=12, ha="left")

plt.savefig("imgs/3_9.png")
```
